```python
# Column from Python list
fuel_economy = [7,5, 9.2, 5.0, 9.6, 8.7]
car_sales["Fuel/100KM"] = fuel_economy
car_sales
```

```python
# So the list must be as long as the dataframe length;
fuel_economy = [7,5, 9.2, 5.0, 9.6, 8.7, 4.7, 5.6, 4.5, 12.2]
#fuel economy has the same length as car_sales
car_sales["Fuel/100KM"] = fuel_economy
car_sales
```

```python
#Example: make column of how much fuel used;
car_sales["Total Fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel/100KM"]
car_sales
```
