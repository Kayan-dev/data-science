##Matplotlib example workflow

```python
# 1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x= [1,2,3,4]
y=[11,22,33,44]
# 3. Setup plot
fig, ax = plt.subplots(figsize=(7.5,5))

# 4. Plot data
ax.plot(x ,y)

# 5. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-label")
# 5. Save & show ( save the figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplots 

```python
# Subplot option 1 
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax0.plot(x, x/2)
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.hist(np.random.randn(1000))
```

```python
# Subplot option 2 
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

### Which to use? Pyplot or matplotlib OO method?
* When plotting something quickly, use pyplot method
* When plotting something more advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target")
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)
# example of how to change x limit;
## ax-set_xlim([45,100])
```

```python
## OO Method form scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50.age,
                    y=over_50.chol,
                    c=over_50.target)
# Customize the plot
ax.set(title="Heart Disease & Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# How to add a horizonta line (mean) in matplotlib plot?
## https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.axhline.html
ax.axhline(over_50.chol.mean(),
          linestyle='--')
```

```python
over_50.target.values
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target)
# customize ax0
ax0.set(title="Heart Disease & Cholesterol Levels",
       xlabel="",
       ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline 
ax0.axhline(y=over_50.chol.mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50.age,
                     y=over_50.thalach,
                     c=over_50.target)
# customize ax0
ax1.set(title="Heart Disease & Max Hearth Rate",
       xlabel="Age",
       ylabel="thalach")
# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="target")
# Add a meanline 
ax1.axhline(y=over_50.thalach.mean(),
           linestyle="--");
# Add Title to (entire) figure
fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

## OO Method form scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50.age,
                    y=over_50.chol,
                    c=over_50.target,
                    cmap="cool") # This changes colour scheme
# Check this link for tutorial 
##  https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py 

# Customize the plot
ax.set(title="Heart Disease & Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
# Change mean() line
ax.axhline(over_50.chol.mean(),
          linestyle='--')
```

```python
# Customizing th y & x axis limitations
plt.style.use('seaborn')

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target,
                     cmap="viridis")
# customize ax0
ax0.set(title="Heart Disease & Cholesterol Levels",
       xlabel="",
       ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline 
ax0.axhline(y=over_50.chol.mean(),
           linestyle="--",
           color="purple");

# Add data to ax1
scatter = ax1.scatter(x=over_50.age,
                     y=over_50.thalach,
                     c=over_50.target,
                     cmap="Set1")
# customize ax1
ax1.set(title="Heart Disease & Max Hearth Rate",
       xlabel="Age",
       ylabel="thalach")
# Change ax1 x-axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])
# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="target")
# Add a meanline 
ax1.axhline(y=over_50.thalach.mean(),
           linestyle="--",
           color="black");
# Add Title to (entire) figure
fig.suptitle("Heart Disease Analysis", fontsize= 16, fontweight="bold");
```
