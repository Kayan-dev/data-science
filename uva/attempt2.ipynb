```python
import string

mess = 'Sample message! Notice: it has punctuation.'

# Check characters to see if they are in punctuation
nopunc = [char for char in mess if char not in string.punctuation]

# Join the characters again to form the string.
nopunc = ''.join(nopunc)
nopunc
```

```python
# Now just remove any stopwords
clean_mess = [word for word in sort if word.lower() not in stopwords.words('english')]
len(clean_mess)
clean_mess[100:200]
```

```python
def text_process(text):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in text if char not in string.punctuation]
    
    # Now just remove any stopwords
    return [word for word in nopunc if word.lower() not in stopwords.words('english')]
```

```python
text_process(clean_mess)
clean_mess
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(clean_mess)

# Print total number of vocab words
print(bow_transformer.vocabulary_)
```
