```python
from urllib import request
url = "https://www.uva.nl/binaries/content/assets/programmas/information-studies/txt-for-assignment-data-science.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
type(raw)
```

```python
from bs4 import BeautifulSoup
raw2 = BeautifulSoup(raw, 'html.parser').get_text()
tokens = word_tokenize(raw2)
tokens[:10]
```

```python
raw_cleaned = nltk.text.get_text(raw)
```

```python
soup = BeautifulSoup(raw)
```

```python
raw_clean = soup.find('text').get_text()
raw_clean[:100]
```

```python
raw3 = BeautifulSoup(raw_clean, 'html.parser').get_text()
tokens = word_tokenize(raw3)
tokens
```

```python
type(tokens)
```

```python
words = [w.lower() for w in tokens]
type(words)
```

```python
vocab = sorted(set(words))
type(vocab)
```

```python
vocab[:30]
```

```python
porter = nltk.PorterStemmer()
porter_words = [porter.stem(t) for t in vocab]
```

```python
lancaster = nltk.LancasterStemmer()
lancaster_words = [lancaster.stem(t) for t in vocab]
```

```python
wnl = nltk.WordNetLemmatizer()
wnl_words = [wnl.lemmatize(t) for t in vocab]
wnl_words
```

```python
type(wnl_words)
```
